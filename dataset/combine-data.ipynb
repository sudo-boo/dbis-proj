```python
import json

# Initialize an empty list to hold all the data
all_data = []

# First, read the initial json file and append its data to all_data
with open('jsons/data.json', 'r') as f:
    data = json.load(f)
    all_data.extend(data)  # use extend to add the data from the first file to the list

# Now, loop through the other json files (data1.json to data15.json) and append their data
for i in range(1, 16):
    with open(f'jsons/data{i}.json', 'r') as f:
        data = json.load(f)
        all_data.extend(data)  # append data from each file

# Finally, write the merged data to 'all-data.json'
with open('all-data.json', 'w') as outfile:
    json.dump(all_data, outfile, indent=4)
```

```python
# read the merged json file
with open('all-data.json', 'r') as f:
    data = json.load(f)
    
print(f"Number of records in all-data.json: {len(data)}")

# sort the data by 'product_id' in ascending order
data.sort(key=lambda x: x['product_id'])

# write the sorted data to 'all-data-sorted.json'
with open('all-data-sorted.json', 'w') as outfile:
    json.dump(data, outfile, indent=4)
```

```python
# convert the json data to a pandas DataFrame
df = pd.DataFrame(data)

df
```

```python
# count the number of times image_urls list is empty
empty_image_urls_count = df['image_urls'].apply(lambda x: len(x) == 0).sum()
print(f"Number of empty image_urls: {empty_image_urls_count}")

# delete the rows where the image_urls list is empty
df = df[df['image_urls'].apply(lambda x: len(x) > 0)]

df
```

```python
# count number of times product_highlights may is empty
empty_product_highlights_count = df['product_highlights'].apply(lambda x: len(x) == 0).sum()
print(f"Number of empty product_highlights: {empty_product_highlights_count}")

# reset the index of the DataFrame
df.reset_index(drop=True, inplace=True)
```

```python
# print the column names of the DataFrame
print("Column names in the DataFrame:")
print(df.columns.tolist())
```

```python
# arrange the columns in the desired order
desired_order = ['product_id', 'name', 'net_qty', 'offer_price', 'discount', 'mrp', 'category', 'image_urls', 'product_highlights', 'product_information', 'search_phrase']
df = df[desired_order]

df
```

```python

# save the cleaned DataFrame to a new CSV file
df.to_csv('all-data-cleaned.csv', index=False)
```
