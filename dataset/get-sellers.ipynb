```python
import pandas as pd
import json
import ast
```

```python
data = pd.read_csv('all-data-cleaned.csv')

data.head()
```

```python
sellers = {}


# Loop through all the rows in the DataFrame
for i in range(len(data)):
    mp = data['product_information'][i]
    
    # Use ast.literal_eval to safely evaluate the string as a dictionary
    mp_dict = ast.literal_eval(mp)
    
    # Now mp_dict is a dictionary, and you can access its elements
    # print(f"Item {i}: {mp_dict}")
    # print(f"Type of Item {i}: {type(mp_dict)}")  # Should now be a dictionary
    seller_license_no = mp_dict.get('Seller License No.', 'NA')
    # print(f"Seller License No.: {seller_license_no}")
    
    seller_name = mp_dict.get('Seller Name', 'NA')
    # print(f"Seller Name: {seller_name}")
    
    seller_address = mp_dict.get('Seller Address', 'NA')
    # print(f"Seller Address: {seller_address}") 
    
    if seller_license_no != 'NA':
        if seller_license_no not in sellers:
            sellers[seller_license_no] = {
                'name': seller_name,
                'address': seller_address,
                'count': 1
            }
        else:
            sellers[seller_license_no]['count'] += 1
```

```python
# print sellers

for key, value in sellers.items():
    print(f"License No.: {key}")
    print(f"Name: {value['name']}")
    print(f"Address: {value['address']}")
    print(f"Count: {value['count']}")
    print()
```
